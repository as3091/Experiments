```python
import sqlite3

# 1. Connect to a database (or create 'example.db')
# Use ':memory:' for a temporary database in RAM
connection = sqlite3.connect('example.db')

# 2. Create a cursor object
cursor = connection.cursor()

# 3. Create a table
cursor.execute('''CREATE TABLE IF NOT EXISTS users 
                  (id INTEGER PRIMARY KEY, name TEXT, age INTEGER)''')

# 4. Insert data using parameterized queries (to prevent SQL injection)
user_data = ('Alice', 30)
cursor.execute("INSERT INTO users (name, age) VALUES (?, ?)", user_data)

# 5. Commit changes and close
connection.commit()
connection.close()
```

```python
connection_obj = sqlite3.connect('example.db')

# cursor object
cursor_obj = connection_obj.cursor()

# to select all column we will use
statement = '''SELECT * FROM users'''

cursor_obj.execute(statement)

print("All the data")
output = cursor_obj.fetchall()
for row in output:
    print(row)

connection_obj.commit()

# Close the connection
connection_obj.close()
```
